```python
import pandas as pd
a=pd.read_excel(r"C:\Users\chauh\Downloads\EDA_Shark_Tank_India.xlsx")
a.head(3)
```

Q1. Find the number of episodes?

```python
b=a['episode_number'].nunique()
print('No. of Episodes are',b)
```

Q2. Max, min , mean of asked amount,asked equity, asked valuation?

```python
a.agg({'pitcher_ask_amount':['max','min','mean'], 'ask_equity':['max','min','mean'], 'ask_valuation':['max','min','mean']})
```

Q3. Max & Min asked- equity,asked-valuation and asked amount episode-wise?

```python
a.groupby('episode_number').agg({'pitcher_ask_amount':['max','min'], 'ask_equity':['max','min'], 'ask_valuation':['max','min']})
```

Q4. Brand names in which 2 ,3 or 4 sharks are invested?

```python
b=a[a['total_sharks_invested'].isin([2,3,4])]
b['brand_name']
```

Q5. Episode wise minimum and maximum of sharks invested?

```python
a.groupby('episode_number').agg({'total_sharks_invested':['max','min']})
```

Q6. Brand names who ask for 1 crore and got a deal?

```python
b=a[(a['total_sharks_invested']>0) & (a['pitcher_ask_amount']==100)]
b['brand_name']
```

Q7. Brand names where deal equity is between 0 to 50?

```python
c=a[(a['deal_equity']>=0) & (a['deal_equity']<=50)]
c['brand_name']
```

Q8. Find the number of brands participated in each episode?

```python
a[['episode_number']].value_counts()
```

Q9. How many sharks participated in this show and What were their names?

```python
b=a.columns[11:18]
b = b.str.replace('_present', '')
c=pd.value_counts(b).count()
print('No. of Sharks are',c)
print('Names=',b)
```
